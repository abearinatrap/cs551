```python
# dataset from https://www.kaggle.com/datasets/open-source-sports/baseball-databank/
import numpy as np
import pandas as pd
```

```python
Players = pd.read_csv("archive/Master.csv")
```

```python
Batting = pd.read_csv("archive/Batting.csv")
print(Batting.head())
for col in Batting.columns:
    print(col, end=" ")

# sb stolen base 
# cs caught stealing
# bb walk
# so strikeout
# ibb intentional walk
# hbp hit by pitch
# sh sacrifice bunt/hit
# sf sacrifice fly
# gidp ground into double play

"""
Ranking/rating system: per year, batting average, on base percentage + slugging with > 30 games played, run + rbi/ gp * 2
graph best individual years,



"""
best_players = []
for index, row in Batting.iterrows():
    try:
        ps = []
        if row['G'] < 30:
            continue

        if row['AB'] < 30:
            continue

        ba = row['H']/row['AB']
        if ba ==1:
            print(row['H'],row['AB'])
        # total bases
        tb = 1*(row['H']-row['2B']-row['3B']-row['HR']) + 2*row['2B'] + 3*row['3B'] + 4*row['HR']

        ops = (row['AB']*(row['H']+row['BB']+row['HBP']) + tb * (row['AB']+row['BB']+row['SF']+row['HBP'])) / (row['AB']*(row['AB']+row['BB']+row['SF']+row['HBP']))
        scoring = (row['R'] + row['RBI'] ) / (2 * row['G'])
        
        final_score = .4 * ba + .2 * ops + .4 * scoring
        #final_score = ba
        
        ps.append(final_score)
        ps.append(row['playerID'])
        ps.append(row['yearID'])
        ps.append(row['teamID'])

        best_players.append(ps)
    except Exception:
        continue
    
print("")
print(len(best_players))
```

```python
best_players.sort()
best_players.reverse()
```

```python
ten_best = []

for i in range(100):
    #print()
    a = best_players[i]
    p = Players.loc[Players['playerID'] == a[1]].iloc[0]
    #print(p.iterrows())
    print(f"Year: {a[2]}, Name: {p['nameGiven']} {p['nameLast']}, Score: {a[0]}")
    ten_best.append({'Score': a[0], 'Year':a[2], 'Name': f"{p['nameGiven']} {p['nameLast']}"})

for i in range(1000,1000):
    #print()
    a = best_players[i]
    p = Players.loc[Players['playerID'] == a[1]].iloc[0]
    #print(p.iterrows())
    #print(f"Year: {a[2]}, Name: {p['nameGiven']} {p['nameLast']}, Score: {a[0]}")
    ten_best.append({'Score': a[0], 'Year':a[2], 'Name': f"{p['nameGiven']} {p['nameLast']}"})
    
```

```python
years = [player['Year'] for player in ten_best]
scores = [player['Score'] for player in ten_best]
names = [player['Name'] for player in ten_best]

# Create a scatter plot
plt.scatter(years, scores, c='blue', marker='o', label='Players')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Score')
plt.title('Baseball Players Performance Over Years')

#fig, ax = plt.subplots()
# Add player names as annotations
for i, name in enumerate(names):
    if scores[i]>=0:
        plt.annotate(name, (years[i], scores[i]), textcoords="offset points", xytext=(0,5), ha='center')

# Show legend
plt.rcParams["figure.figsize"] = (10,10)
```
